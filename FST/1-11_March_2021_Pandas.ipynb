```python
import pandas as pd
d={"roll":[1,2,3,4,5],"name":["a","b","c","d","e"],"marks":[11,13,12,15,16]}
        
```

```python
df=pd.DataFrame(d)
print(df)

#iloc

#dataframe.iloc[row,col]

print(df.iloc[1,1])
```

```python
print(df.iloc[[1,3],[0,1,2]])
```

```python
print(df.iloc[[1,3],:])
```

```python
print(df.iloc[1::2,:])
```

```python
#loc

#dataframe.loc[row,col]
#row=>index
#col=>column name


print(df.loc[1,"name"])
print(df.loc[1::2])
print(df.loc[[1,2,3],["name","marks"]])
```
