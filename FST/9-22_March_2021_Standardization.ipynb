```python
#Standardization
import pandas as pd 
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
data=np.array([1.3,2.1,4.5,6.5])
data
mean=np.mean(data)
std=np.std(data)
print(np.mean(data))
print(np.std(data))
```

```python
#data-mean/std
scaled_data=(data-mean)/std
print("after scaling")
print(scaled_data)
print(np.mean(scaled_data))
print(np.std(scaled_data))
```

# Scaling

```python
data=np.array([1.3,2.1,4.5,6.5])
data

dmin=np.min(data)
dmax=np.max(data)

# x-xmin/xmax-xmin formula
print((data-dmin)/(dmax-dmin))
```

```python
data=data.reshape(-1,1)
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
m.fit(data)
print(m.transform(data))
```

```python
s=np.array([0,0.15384615,0.61538462,1])
s
mean=np.mean(s)
std=np.std(s)
print(np.mean(s))
print(np.std(s))
```

```python
#data-mean/std
scaled_data=(s-mean)/std
print("after scaling")
print(scaled_data)
print(np.mean(scaled_data))
print(np.std(scaled_data))
```
