# Distribution

# Uniform Distribution

```python
import seaborn as sns
from scipy.stats import uniform
n=10000
start=10
width=20
data_uniform=uniform.rvs(size=n,loc=start,scale=width)
ax=sns.histplot(data_uniform,
                bins=10,
                kde=True,
                color="skyblue",
               )
ax.set(xlabel="Uniform Distribution",ylabel="Frequency")
```

```python
import seaborn as sns
from scipy.stats import uniform
n=10000
start=10
width=20
data_uniform=uniform.rvs(size=n,loc=start,scale=width)
ax=sns.distplot(data_uniform,
                bins=10,
                kde=True,
                color="skyblue",
               )
ax.set(xlabel="Uniform Distribution",ylabel="Frequency")
```

```python
import seaborn as sns
from scipy.stats import uniform
n=10000
start=10
width=20
data_uniform=uniform.rvs(size=n,loc=start,scale=width)
ax=sns.displot(data_uniform,
                bins=10,
                kde=True,
                color="skyblue",
               )
ax.set(xlabel="Uniform Distribution",ylabel="Frequency")
```

# Normal Distribution

```python
import seaborn as sns
from scipy.stats import norm
data_normal=norm.rvs(size=10000,loc=0,scale=1)
ax=sns.histplot(data_normal,kde=True,color="red",)
ax.set(xlabel="Normal Distribution",ylabel="Frequency")
```

# Gamma Disribution

```python
import seaborn as sns
from scipy.stats import gamma
data_gamma=gamma.rvs(a=5,size=10000)
ax=sns.histplot(data_gamma,kde=True,bins=100,color="red",)
ax.set(xlabel="Gamma Distribution",ylabel="Frequency")
```

# Exponential Distribution

```python
import seaborn as sns
from scipy.stats import expon
data_expon=expon.rvs(scale=1,loc=0,size=10000)
ax=sns.histplot(data_expon,kde=True,bins=100,color="red",)
ax.set(xlabel="Gamma Distribution",ylabel="Frequency")
```

# Poission Distribution

```python
import seaborn as sns
from scipy.stats import poisson
data_poisson=poisson.rvs(mu=3,size=10000)
ax=sns.histplot(data_poisson,kde=False,bins=30,color="blue",)
ax.set(xlabel="poisson Distribution",ylabel="Frequency")
```

# Binomial Distribution

```python
import seaborn as sns
from scipy.stats import binom
data_binom=binom.rvs(n=10,p=0.8,size=10000)
ax=sns.histplot(data_binom,kde=False,color="blue",)
ax.set(xlabel="Binomial Distribution",ylabel="Frequency")
```
