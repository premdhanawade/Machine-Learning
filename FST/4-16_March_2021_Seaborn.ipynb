```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("iris")
df.head()
```

# Scatter plot

```python
#scatter plot
sns.scatterplot(x="species",y="sepal_width",data=df)
plt.show
```

# Swarmplot

```python
df=sns.load_dataset("iris")
sns.swarmplot(x="species",y="sepal_width",data=df)
plt.show
```

# Barplot

```python
# barplot
sns.barplot(x="sepal_length",y="sepal_width",data=df)
```

# Histogram

```python
#sns hist
sns.histplot(x="sepal_width",data=df)
```

# Relplot

```python
sns.relplot(x="sepal_width",y="sepal_length",data=df)
```

# Lineplot

```python
sns.lineplot(x="sepal_width",y="sepal_length",data=df)
```

# Regplot

```python
sns.regplot(x="sepal_width",y="sepal_length",data=df)
```

# Jointplot

```python
# joint between hist & scatter
sns.jointplot(x="sepal_width",y="sepal_length",data=df)
```

```python
sns.jointplot(x="sepal_width",y="sepal_length",data=df,kind="hex")
```

# Boxplot

```python
sns.boxplot(x="sepal_width",y="sepal_length",data=df)
```

# Violinplot

```python
sns.violinplot(x="sepal_width",y="sepal_length",data=df)
```

```python
sns.violinplot(x="sepal_width",data=df)
```

# Factorplot

```python
sns.factorplot("species",data=df,kind="count")
```
